```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
apple = pd.read_csv("appstore_games.csv")

apple.describe()
```

```python
apple
```

# Price

```python
free = apple[apple["Price"]==0].count()[0]

plt.style.use("ggplot")

plt.figure(figsize=(10, 8))

plt.subplot(131)
price = [free, apple.count()[0]-free]
labels = ["Free", "Not-Free"]
plt.pie(price, labels=labels, autopct="%.2f %%", pctdistance=0.5)
plt.title("Free and Not-Free apps %")
plt.tight_layout()

plt.subplot(132)
notfree = apple[(apple["Price"] > 0) & (apple["Price"] < 20)]
plt.hist(notfree["Price"], bins=np.arange(0,21, 2))
plt.title("Not-Free apps price")
plt.ylabel("Amount of apps")
plt.xlabel("Price")
plt.tight_layout()

plt.show()
```
