```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pokemon = pd.read_csv("pokemon.csv")

pokemon
```

# Generation #

```python

mingen = pokemon["generation"].min()
maxgen = pokemon["generation"].max()

countgen = []
labels = []
for i in range(mingen, maxgen+1):
    countgen.append(pokemon[pokemon["generation"] == i].count()[0])
    labels.append("Gen " + str(i))

print(countgen)
plt.style.use("ggplot")
plt.figure(figsize=(10, 8))

plt.pie(countgen, labels=labels, autopct="%.2f %%", pctdistance=0.8)

plt.title("Number of pokemons per generation")
plt.show()

```
